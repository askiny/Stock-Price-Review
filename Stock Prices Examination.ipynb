# Stock Prices Examination

In this project, I will examine stock prize data of ten companies between 2019 and 2022 using YahooFinance data.

(Alphabet (Google), Meta (Facebook), Cisco, Intel, AMD, Qualcomm, Apple, Samsung, Xiaomi, and Tesla.)

- I will create a data frame named Change that can take the values 1, 0 and -1. If the change between the opening and closing prices of a day is -2.5%, the Change will be -1. If this change is 2.5%, Change will be 1. If the change is between -2.5% and 2.5%, Change will appear as 0.

- Identifying the longest continuous time intervals during which the CHANGE consistently remains 1 and -1 for each company.

- Creating another pandas data frame Difference consisting of values -1, 0, 1 for wach day within the specifies time period. If the opening price of a day is within -2.5% and 2.5% of the previous day's opening price, I will assign the Change value 0. If the change in the opening price in comparison to the previous day is greater than 2.5% I will assign the value of 1. If the change in the opening price is less than -2.5% I will assign the Cahnge value of -1.

- Determining the longest continuous time intervals which the Difference consistently remains 1 and -1 for each company.

```python
import yfinance as yf
import pandas as pd
```

I created a list of companies whose data I would use. Using the elements of the list I created, I added the date range I wanted and then pulled the data. I used the "rounding = True" function to make the decimal numbers appear more clear when pulling the data and grouped the data according to companies.

```python
companies = ["GOOG", "F", "CSCO", "INTC", "AMD", "QCOM", "AAPL", "005930.KS", "1810.HK", "TSLA"]

data = yf.download(companies, start = '2019-1-1', end = '2022-12-31', rounding = True, group_by = "company")
data['Date'] = data.index

data
```

The percentage_change variable allows us to keep the difference between the company's opening price and closing price.

I defined 2.5% as the threshold value. I will use this variable when defining the change dataframe.

While creating the change dataframe, I assigned the value 1 to the companies that showed an increase in the values exceeding the threshold value, and -1 to the companies that showed a decrease in the values exceeding the threshold value. If the change experienced by the company does not exceed the threshold value, I gave that company a value of 0. I repeated this process for all companies using a for loop.

The "lambda x:" variable I used when creating the change dataframe allowed me to complete the operations without an extra variable kept in memory.

Then I printed the change dataframe to the screen to see if it worked properly.

```python
percentage_change = pd.DataFrame()
change = pd.DataFrame()

threshold = 0.025

for company in companies:
    percentage_change[company] = (data[company]['Close'] - data[company]['Open']) / data[company]['Open']
    change[company + '_change'] = percentage_change[company].apply(
        lambda x: 1 if x > threshold else ( -1 if x < -threshold else 0)
    )
    
change
```

First, I pulled the opening prices of the companies and defined a variable called previous_opening_prices to pull the opening price of the day before the mentioned date.

Then, I defined the variable percentage_opening_change to calculate the difference.

Then I created the difference dataframe and defined the dataframe to not start from the first column to exclude the first day, 01.01.2019.

Next, I wrote a for loop that calculates one day's opening price and the previous day's opening price for all companies.

I added the data I obtained from the for loop to the DIFFERENCE dataframe. I defined increasing, decreasing and remaining constant intervals for the threshold value I determined before. I did this again using "lambda x:".

Then I arranged the columns and printed them to the screen to see if the code worked properly.

```python
opening_prices = yf.download(companies, start = '2019-1-1', end = '2022-12-31')["Open"]
previous_opening_prices = opening_prices.shift(1)

percentage_opening_change = (opening_prices - previous_opening_prices) / previous_opening_prices

difference = pd.DataFrame((data)['Date'][1:])
for company in companies:
    price_diff = opening_prices[company].diff()
    
    difference[company + '_difference'] = price_diff.apply(
        lambda x: 1 if x > threshold else (-1 if x < -threshold else 0)
    )

difference = difference[['Date'] + [f"{company}_difference" for company in companies]]

difference.drop(columns = 'Date', inplace = True)

difference
```

And now, I will work on finding longest intervals for change and difference dataframes:

This function finds the longest interval containing a given value (1 or -1) in the 'change' or 'difference' DataFrames. The first loop traverses the rows on the DataFrame for each date, and the second loop checks the change/difference columns for specific companies. If a particular company's change/difference column contains the desired value, it saves it as a interval. If not, it adds the current range to the longest range by comparing lengths and resets the current range.

```python
def find_longest_intervals(df, name, value):
    longest_intervals = {}
    current_intervals = {}
    
    for index, row in df.iterrows():
        date = index
        for company in companies:
            column_name = f"{company}_{name}"
            if row[column_name] == value:
                if company not in current_intervals:
                    current_intervals[company] = {'Start': date, 'End': date}
                else:
                    current_intervals[company]['End'] = date
            else:
                if company in current_intervals:
                    if company not in longest_intervals:
                        longest_intervals[company] = {'Start': current_intervals[company]['Start'], 'End': current_intervals[company]['End']}
                    else:
                        if current_intervals[company]['End'] - current_intervals[company]['Start'] > longest_intervals[company]['End'] - longest_intervals[company]['Start']:
                            longest_intervals[company] = {'Start': current_intervals[company]['Start'], 'End': current_intervals[company]['End']}
                    del current_intervals[company]
    df_longest_intervals = pd.DataFrame(longest_intervals).T
    return df_longest_intervals
```

Now I will print the longest interval tables on screen:

```python
print("Longest Consecutive Intervals with Difference Value 1:")
find_longest_intervals(difference, 'difference', 1)
```

```python
print("Longest Consecutive Intervals with Difference Value -1:")
find_longest_intervals(difference, 'difference', -1)
```

```python
print("Longest Consecutive Intervals with Change Value 1:")
find_longest_intervals(change, 'change', 1)
```

```python
print("Longest Consecutive Intervals with Change Value -1:")
find_longest_intervals(change, 'change', -1)
```
